```python
# Simulator adding in diseases for biometrics to test whether nodes work
diseases = {'nodes':[], 'links':[]}
num_dis = 80
last_dis = -1
for i in range(num_dis):
    dis_name = 'disease_'+ str(i)
    diseases['nodes'].append({'name': dis_name, 'group': i})
    if last_dis != -1:
        diseases['links'].append({'source': last_dis, 'target': i, 'value': 50})
        last_dis = i
    else:
        last_dis = i

diseases
```

```python
counter = 0
nodes = []
links = []

for biom in list(processed_dict_bio.keys()):
    biom_name = {}
    biom_name['name'] = biom
    biom_name['group'] = list(processed_dict_bio.keys()).index(biom) #put sth else here
    nodes.append(biom_name)
    for conn in processed_dict_bio[biom]:
        target = list(processed_dict_bio.keys()).index(conn[0])
        no_flag = False
        for link in links:
            if link["source"] == target:
                if link["target"] == counter:
                    no_flag = True
        if no_flag == True:
            continue
        biom_dict = {}
        biom_dict["source"] = counter
        biom_dict["target"] = target
        biom_dict["value"] = conn[1]
        links.append(biom_dict)
    counter = counter + 1
    
data = {
    "nodes": nodes,
    "links": links
}

data
```
